```python
# find the min,max,sum,mean,median,std and dot product
print("minimum value=",np.min(ran))
print("min value=",ran.min())
print("max value=",np.max(ran))
print("max value=",ran.max())
print("sum=",np.sum(ran))
print("sum=",ran.sum())
print("mean=",np.mean(ran))
print("mean=",ran.mean())
print("median=",np.median(ran))
print(ran.dot(man))
print(ran.std())
```

```python
# difference b/w user defined func and np.vectorize fun
def greater(a,b):
    if a>b:
        return a
    else:return b
    
greater(10,34),greater(35,12),greater([45,23,10],[67,35,20])
```

```python
g=np.vectorize(greater)
g([45,23,10],[67,35,20])
```
